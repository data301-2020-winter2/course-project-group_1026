```python
import pandas as pd
import numpy as np
```

```python
def load_and_process(url_or_path_to_csv_file):
    

    df1 = (
        pd.read_csv(url_or_path_to_csv_file)
        .dropna()
        .drop(['VotingAgeCitizen','Office','Drive','Carpool','Transit','Walk','Service','Construction','Hispanic','White','Black','White','Native','Production','OtherTransp','MeanCommute','Asian','Pacific','Professional','WorkAtHome'], axis ="columns")
    )

    df2 = (
        df1
        .loc[(df1['State'] == 'Florida') | (df1['State'] == 'Texas')|(df1['State'] == 'California')|(df1['State'] == 'New York')]
        .reset_index()
        .drop(['index'], axis=1) 
    )
  
    return df2
```

```python
data = load_and_process("acs2017_county_data.csv")
data.head()
```

```python
data.to_csv(r'processed.csv', index = None)
```
